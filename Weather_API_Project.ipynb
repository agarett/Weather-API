```python
# Dependencies
import csv 
import matplotlib.pyplot as plt
import openweathermapy as owm
import pandas as pd
from citipy import citipy
import random
import numpy

# import api_key from config file
from config import api_key
```

```python
#get a random list of cities from citipy to pull data from 
cities = []
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "&units=imperial&q="

while len(cities) < 700:
    latitude = round(random.uniform(-90.00, 90.00), 2)
    longitude = round(random.uniform(-180.00, 180.00), 2)
    city = citipy.nearest_city(latitude, longitude)
    if city not in cities:
        cities.append(city.city_name + "," + city.country_code.upper())
    else:
        continue
        
```

```python
#create a dataframe for all of the necessary weather values 

cities_df = pd.DataFrame({'City':cities})
cities_df["Latitude"] = ""
cities_df['Temperature'] = ""
cities_df["Humidity"] = ""
cities_df['Cloudiness'] = ""
cities_df['Wind Speed'] = ""
```

```python
weather_data = []
city_number = 1
settings = {"units": "imperial", "APPID": api_key}

#Iterate through the rows in the df and append values from the weather API 

for index, row in cities_df.iterrows(): 
    city = row['City']
    try:  
        current_city = [owm.get_current(city, **settings)]  
    
        cities_df.loc[index, "Latitude"] = current_city[0]['coord']['lat']
        cities_df.loc[index, "Temperature"] = current_city[0]['main']['temp_max']
        cities_df.loc[index, "Humidity"] = current_city[0]['main']['humidity']
        cities_df.loc[index, "Cloudiness"] = current_city[0]['clouds']['all']
        cities_df.loc[index, "Wind Speed"] = current_city[0]['wind']['speed']
    except:
        print("City not found")
        pass
        
    #print out each city that is called
    
    print(f"Processing record {city_number}: {city}.")
    print(f"{url}{units}{city}")
    city_number += 1    
```

```python
#convert df values to numeric so they can be plotted; then save as csv

cities_df = cities_df.apply(pd.to_numeric, errors='ignore')
cities_df.to_csv("WeatherAPI.csv", encoding='utf-8')
```

```python
#plot the df comparing latitude against temperature

fig1 = cities_df.plot(kind="scatter", x="Latitude", y="Temperature", grid=True, figsize=(10,5),
              title="City Latitude vs. Temperature (06/22/2018)")

plt.ylim(-100, 150)
plt.xlim(-80, 100)
plt.grid(True)
plt.savefig("weather_scatter1.png")
plt.show()
```
